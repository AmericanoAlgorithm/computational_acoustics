# Two-Dimensional FEM Example Meshes

This notebook shows some examples for two-dimensional meshes as used in the [Finite Element Method](https://en.wikipedia.org/wiki/Finite_element_method) (FEM) for spatial discretization. The meshes are generated by [FEniCS](https://fenicsproject.org/), an open-source framework for numerical solution of PDEs.

```python
from dolfin import *
import mshr
%matplotlib inline
```

```python
mesh = RectangleMesh(Point(-2,-2), Point(2, 2), 20, 20, "right/left")
plot(mesh);
```

```python
domain = mshr.Rectangle(Point(-4,-4), Point(4,4))
mesh = mshr.generate_mesh(domain, 20)
plot(mesh);
```

```python
domain = mshr.Rectangle(Point(0, 0), Point(3,4)) + mshr.Rectangle(Point(3, 1.5), Point(3.5, 2.5)) + mshr.Rectangle(Point(3.5, 0), Point(6, 4))
mesh = mshr.generate_mesh(domain, 20)

plot(mesh);
```

```python
domain = mshr.Rectangle(Point(-4,-4), Point(4,4)) - mshr.Rectangle(Point(-.75,-.75), Point(.75,.75))
mesh = mshr.generate_mesh(domain, 20)
plot(mesh);
```

```python
domain = mshr.Rectangle(Point(-4,-4), Point(4,4))-mshr.Circle(Point(0,0), 1)
mesh = mshr.generate_mesh(domain, 20)
plot(mesh);
```

```python
domain_vertices = [Point(0.0, 0.0),
                  Point(5.0, 0.0),
                  Point(5.0, 0.75),
                  Point(4.5, 0.75),
                  Point(4.0, 1.50),
                  Point(2.0, 1.50),
                  Point(1.0, 0.75),
                  Point(0.0, 0.75),
                  Point(0.0, 0.0)]

domain = mshr.Polygon(domain_vertices)
mesh = mshr.generate_mesh(domain, 20)

plot(mesh);
```
